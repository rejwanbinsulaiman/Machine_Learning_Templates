## Part 1 - Data Preprocessing

### Importing the dataset

```python
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

### Encoding categorical data

Label Encoding the "Gender" column

```python
from sklearn.preprocessing import LabelEncoder
labelEncoer_X_1 = LabelEncoder()
X[:, 1] = labelEncoer_X_1.fit_transform(X[:, 1])
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoer_X_2 = LabelEncoder()
X[:, 2] = labelEncoer_X_2.fit_transform(X[:, 2]) # encoding counrty
```

```python
X
```

One Hot Encoding the "Geography" column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ColumnTransformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough') 
# [1] is the column index number
X = np.array(ColumnTransformer.fit_transform(X))
```

```python
X = X[:, 1:] # so can avoid dummy variabe trap, by removing one colum if there are three /4/5
```

```python
print(X[0])
```

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

## Part 2 - Building the ANN

```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense
```

### Initializing the ANN

```python
tf.__version__
```

```python
ann = tf.keras.models.Sequential() #classifier = squencial()
```

### Adding the input layer and the first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu', input_dim =11)) 
#unit = 6, is the neural layers (its calculated from the avg of total features and the output (predicting - y ) neuro)
# we need to use the input_dim param as there are no layers that can say how many layers to expect
```

### Adding the second hidden layer

```python

ann.add(tf.keras.layers.Dense(units=6, activation='relu')) 
#we do not need , input_dim = param as the NN knows how many inputs expected.
```

### Adding the output layer

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

## Part 3 - Training the ANN

### Compiling the ANN

```python
# ann = tf.keras.models.Sequential() #classifier = squencial()
# ann.add(tf.keras.layers.Dense(units=6, activation='relu', input_dim =11)) 
# ann.add(tf.keras.layers.Dense(units=6, activation='relu')) 
# ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
# ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
#all together
```

```python

ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
# if the dependent variable has catagorical outcome (predicting variable) , which is 2/3 then the 
# loss = categorical_crossentropy
```

### Training the ANN on the Training set

```python
ann.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

### Predicting the Test set results

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

**Important note 1:** Notice that the values of the features were all input in a double pair of square brackets. **Important note 2:** Notice also that the "France" country was not input as a string in the last column but as "1, 0, 0" in the first three columns. And be careful to include these values in the first three columns, because the dummy variables are always created in the first columns.

# Evaluating the ANN using scikit learn (k-Fold cross validation)

```python
#as the ann is in keras and the k fold validation is in sk learn we need a wraper to combine both libraray
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense
```

```python

def build_classifier():
    ann = tf.keras.models.Sequential() #classifier = squencial()
    ann.add(tf.keras.layers.Dense(units=6, activation='relu', input_dim =11)) 
    ann.add(tf.keras.layers.Dense(units=6, activation='relu')) 
    ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
    ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return ann
ann = KerasClassifier(build_fn= build_classifier, nb_epoch=100, batch_size=10)
accuracies = cross_val_score(estimator=ann, X = X_train ,y =y_train, cv = 10 ,n_jobs=-1)
```

# Parameter Tuning

### Drop out regularization to reduce overfitting if needed
##### BEFRE TUNING PARAMETER WE NEED TO DEAL WITH THE OVERFITING PROBLEM


```python
from keras.layers import Dropout
ann = tf.keras.models.Sequential() #classifier = squencial()

ann.add(tf.keras.layers.Dense(units=6, activation='relu', input_dim =11)) 
ann.add(Dropout( 0.1)) 

ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(Dropout(0.1)) 
# IF THE MODEL IS OVER FITTED WE NEEED TO TURN SOME RANDO, NEURONS OFF SO THAT NEURONS DONOT
#BE DEPENDET ON EACH OTHER TOO MUCH [P = 0.1 MEANS ONE PARCENT ] [WE CAN INCREASE P = 1-5 NOT MORE THAN 0.5 AS IT MIGHT TURN A
#ALL THE NEURONS]
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
    #PARAMETER TUNING
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense
def build_classifier(optimizer):
        ann = tf.keras.models.Sequential() #classifier = squencial()
        ann.add(tf.keras.layers.Dense(units=6, activation='relu', input_dim =11)) 
        ann.add(tf.keras.layers.Dense(units=6, activation='relu')) 
        ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
        ann.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
        return ann
ann = KerasClassifier(build_fn= build_classifier)
parameters = {'batch_size' : [25,32],
              'epochs' : [100,500],
              'optimizer' : ['adam','rmsprop']
             }
gridsearch = GridSearchCV(estimator=ann,
            param_grid = parameters,
                          scoring= 'accuracy',
                          cv=10
                         )
```

```python
gridsearch = gridsearch.fit(X_train,y_train)
```

```python
best_parameter = gridsearch.best_params_
best_accuracy = gridsearch.best_score_
```
